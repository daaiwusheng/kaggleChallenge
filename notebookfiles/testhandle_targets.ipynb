```python
import numpy as np
import os, sys
dir2 = os.path.abspath('')
dir1 = os.path.dirname(dir2)
if not dir1 in sys.path: sys.path.append(dir1)
import matplotlib.pyplot as plt # To install: pip install matplotlib
from utility.rle_tool import *
from utility.tool import *
from dataprocess.write_patches import *
from PIL import Image
from skimage.morphology import remove_small_holes
from models.dataset import *
```

```python
dataprovider = KaggleDataSaver(256)
# visualise some examples from the dataset 
number = 3
fig, ax = plt.subplots(3,number)
for i, ax in enumerate(ax.flatten()):
    if i < 6:
        if i < number:
            image_name = dataprovider.train_images[i]
        elif i > 2 and i < 6:
            image_name = dataprovider.train_labels[i-number]
        plottable_image = Image.open(image_name).convert("L")
        image_array = np.asarray(plottable_image)
        ax.imshow(plottable_image, cmap='gray_r')
    else:
        image_name = dataprovider.train_labels[i-2*number]
        plottable_image = Image.open(image_name).convert("L")
        image_array = np.asarray(plottable_image)
        image_array = remove_small_holes(image_array, 2) 
        plottable_image = Image.fromarray(image_array)
        ax.imshow(plottable_image, cmap='gray_r')
```
